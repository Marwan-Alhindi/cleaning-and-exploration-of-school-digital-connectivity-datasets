```python
# We will use these functions for all numerical attributes.

def to_float(column):
    """Converts a column to float"""
    all_data_copy[column] = all_data_copy[column].str.replace('%','')
    all_data_copy[column] = all_data_copy[column].astype(float)
    all_data_copy[column] = all_data_copy[column]/100

def impossible_values(column):
    """Checks if there are any impossible values in a column"""
    errors = []
    for cell in all_data_copy[column]:
        if cell > 1 or cell < 0:
            errors.append(cell)
            print(cell) # Let us see which observation is this
    return errors
```

```python
# Let us first change the data type from object to float
to_float('Total')

all_data_copy['Total']
```

```python
# Data entry or impossible values
impossible_values('Total')

# Let us see which observation is this
all_data_copy[all_data_copy['Total'] == 1.79]
```

```python
# To deal with it, let us see all Ukraine observations
all_data_copy[all_data_copy['Countries and areas'] == 'Ukraine']
# We can see that the total is the value of Urban (Residence). Let us replace 1.79 to 0.79 because 1 seems to be data entry error.
```

```python
# Replacing 1.79 to 0.79
all_data_copy['Total'] = all_data_copy['Total'].replace(1.79,0.79)
```
